#### Preprocess the data

```python
import re, string, unicodedata
import nltk
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
```

```python
"""Getting sentences from the corpus"""
sents = nltk.sent_tokenize(corpus)

def to_lowercase(words):
    """Convert all characters to lowercase from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = word.lower()
        new_words.append(new_word)
    return new_words

def remove_punctuation(words):
    """Remove punctuation and _ from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = re.sub(r'[^\w\s]', '', word)
        if new_word != '' and new_word != '_':
            if new_word[0] == '_':
                  new_word = new_word[1:]
            if new_word[-1] == '_':
                new_word = new_word[:-2]
            new_words.append(new_word) 
    return new_words

def remove_non_ascii(words):
    """Remove non-ASCII characters from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_words.append(new_word)
    return new_words

def normalize(words):
    words = remove_non_ascii(words)
    words = to_lowercase(words)
    words = remove_punctuation(words)
    return words

preprocessed_sents = []
for sent in sents:
    words = nltk.word_tokenize(sent)
    words = normalize(words)
    new_sent = " ".join(words)
    preprocessed_sents.append(new_sent)

print(len(preprocessed_sents))
preprocessed_sents[0]
```

```python
#Write code
from nltk.util import ngrams
unigrams = []
bigrams = []
trigrams = []
fourgrams = []

def get_ngrams(text, n):
    n_grams = ngrams(word_tokenize(text), n)
    return [ ' '.join(grams) for grams in n_grams]

for sent in preprocessed_sents:
    unigrams = unigrams + get_ngrams(sent, 1)
    bigrams = bigrams + get_ngrams(sent, 2)
    trigrams = trigrams + get_ngrams(sent, 3)
    fourgrams = fourgrams + get_ngrams(sent, 4)
    
print(bigrams)
```

```python
stopwords = nltk.corpus.stopwords.words('english')


#print top 10 unigrams, bigrams after removing stopwords without smoothning
uni_processed = [p for p in unigrams if p not in stopwords]
unigrams_fdist = nltk.FreqDist(uni_processed)

bi_processed = [p for p in bigrams if not all(word in stopwords for word in p.split())]
bigrams_fdist = nltk.FreqDist(bi_processed)

tri_processed = [p for p in trigrams if not all(word in stopwords for word in p.split())]
trigrams_fdist = nltk.FreqDist(tri_processed)

four_processed = [p for p in fourgrams if not all(word in stopwords for word in p.split())]
fourgrams_fdist = nltk.FreqDist(four_processed)

#print top 10 bigrams, trigrams, fourgrams after removing stopwords
print("******Top 10 unigrams******")
dic_top_unigrams = {}
for word, frequency in unigrams_fdist.most_common(10):
    dic_top_unigrams[word] = frequency
print(dic_top_unigrams)

print("******Top 10 bigrams******")
dic_top_bigrams = {}
for word, frequency in bigrams_fdist.most_common(10):
    dic_top_bigrams[word] = frequency
print(dic_top_bigrams)

print("******Top 10 trigrams******")
dic_top_trigrams = {}
for word, frequency in trigrams_fdist.most_common(10):
    dic_top_trigrams[word] = frequency
print(dic_top_trigrams)

print("******Top 10 fourgrams******")
dic_top_fourgrams = {}
for word, frequency in fourgrams_fdist.most_common(10):
    dic_top_fourgrams[word] = frequency
print(dic_top_fourgrams)
```

# Applying Smoothing


Assume additional training data in which each possible N-gram occurs exactly once and adjust estimates.

> ### $ Probability(ngram) = \frac{Count(ngram)+1}{ N\, +\, V} $

N: Total number of N-grams <br>
V: Number of unique N-grams



```python
#print top 10 unigrams, bigrams after removing stopwords and with smoothning

bi_processed = [p for p in bigrams if not all(word in stopwords for word in p.split())]
bigrams_fdist = dict(nltk.FreqDist(bi_processed))

tri_processed = [p for p in trigrams if not all(word in stopwords for word in p.split())]
trigrams_fdist = dict(nltk.FreqDist(tri_processed))

four_processed = [p for p in fourgrams if not all(word in stopwords for word in p.split())]
fourgrams_fdist = dict(nltk.FreqDist(four_processed))

dic_top_bigrams = {}
for item in bigrams_fdist.keys():
    dic_top_bigrams[item] = (bigrams_fdist[item]+1)/(len(bigrams)+len(bigrams_fdist))
    
dic_top_trigrams = {}
for item in trigrams_fdist.keys():
    dic_top_trigrams[item] = (trigrams_fdist[item]+1)/(len(trigrams)+len(trigrams_fdist))

dic_top_fourgrams = {}
for item in fourgrams_fdist.keys():
    dic_top_fourgrams[item] = (fourgrams_fdist[item]+1)/(len(fourgrams)+len(fourgrams_fdist))
    
#print top 10 bigrams, trigrams, fourgrams after removing stopwords
print("******Top 10 bigrams******")
print(sorted(dic_top_bigrams.items(), key = lambda kv:(kv[1], kv[0]), reverse=True)[:10]) 

print("******Top 10 trigrams******")
print(sorted(dic_top_trigrams.items(), key = lambda kv:(kv[1], kv[0]), reverse=True)[:10]) 

print("******Top 10 fourgrams******")
print(sorted(dic_top_fourgrams.items(), key = lambda kv:(kv[1], kv[0]), reverse=True)[:10]) 
```
